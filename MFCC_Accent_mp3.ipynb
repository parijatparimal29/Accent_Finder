```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import statistics as stats
import IPython.display as ipd
import csv
import random
import warnings
warnings.simplefilter('ignore')
```

```python
input_csv = []
test_csv = []
skip = True
with open('audio_limited.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in csvreader:
        if(skip):
            skip = False
            continue
        temp = []
        for col in row:
            temp = col.split(',')
            accent = temp[1]
            file = temp[0]
        if(random.randint(1,10)%7!=0):
            input_csv.append([accent,file])
        else:
            test_csv.append([accent,file])


input = []
test = []
for row in input_csv:
    y, sr = librosa.load("audio/"+row[1]+".mp3")
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=12)
    mfcc = []
    for m in mfccs:
        mfcc.append(stats.mean(m))
    input.append([row[0],mfcc])
for row in test_csv:
    y, sr = librosa.load("audio/"+row[1]+".mp3")
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=12)
    mfcc = []
    for m in mfccs:
        mfcc.append(stats.mean(m))
    test.append([row[0],mfcc])
```

```python
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

base = []
acc = []
for i in range(1):
    random.shuffle(input)
    random.shuffle(test)
    Y = [i[0] for i in input]
    X = [(i[1]) for i in input]
    for x in X:
        for i in range(0,12):
            x[i] = float(x[i])
    test_Y = [i[0] for i in test]
    test_X = [(i[1]) for i in test]
    for t_x in test_X:
        for i in range(0,12):
            t_x[i] = float(t_x[i])

    X_np = np.array(X)
    Y_np = np.array(Y)
    test_X_np = np.array(test_X)
    test_Y_np = np.array(test_Y)

    clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
    clf.fit(X_np, Y_np)

    pred = clf.predict(test_X_np)
    print("Prediction |   Actual   | Pred = Act?")
    print("-------------------------------------")
    correct = []
    for i in range(0,len(pred)):
        correct.append(pred[i]==test_Y[i])
        print(pred[i], "      |",test_Y_np[i],"\t|", pred[i] == test_Y_np[i])
    accuracy = correct.count(True)/len(correct)
    uniq = set(test_Y)
    baseline = 0
    for u in uniq:
        if(test_Y.count(u)>baseline):
            baseline = test_Y.count(u)
    baseline /= len(correct)
    print("Baseline: ",baseline*100)
    print("Accuracy:",accuracy*100)
    base.append(baseline*100)
    acc.append(accuracy*100)

    #print("-------------------------------------")
    #print("Baseline: ",baseline,"\nAccuracy: ",accuracy)
#plt.plot(acc)
#plt.plot(base)
#plt.grid()
```
